### 반복문 예제 문제 (다중 for문)
- 페르마의 마지막 정리가 참인가 확인

> a<sup>n</sup> + b<sup>n</sup> = c<sup>n</sup> 이 식에서 n이 2보다 큰 자연수라면 a, b, c에 들어갈 수 있는 자연수는 존재하지 않는다. 

- n이 3이라 가정을 하고 a, b, c에 1부터 10까지 값들을 넣어서 해당하는 조건식이 참인 경우가 존재하는가를 확인 

```python
n = 3
_end = 10
# a가 1부터 10까지 반복을 하는 반복문을 생성
for a in range(1, _end+1, 1):
    # b가 1부터 10까지 반복을 하는 반복문을 생성 
    for b in range(1, _end+1, 1):
        # c가 1부터 10까지 반복을 하는 반복문을 생성 
        for c in range(1, _end+1, 1):
            # 마지막 정리 공식(a**n + b **n == c**n)이 참인가? 조건문 생성
            if a**n + b**n == c**n:
                # 조건식이 참인 경우 a,b,c 출력
                print(a, b, c)
# 프로그램 종료 출력
print('프로그램 종료')
```

### 반복문 문제 
- 소수인 데이터를 출력하라 
- 1부터 10까지의 소수만 출력하시오
    - 5 -> 2부터 시작해서 4까지 5와 나눠서 나머지가 0이 있는가?

```python
## 2부터 10까지 반복을 하는 반복문을 생성
cnt = 0 
for i in range(2, 5001, 1):
    # 변수(check)를 생성 하여 값을 True
    check = True
    ## 반복문 생성 2부터 i - 1까지 반복을 하는 반복문을 생성
    for j in range(2, i, 1):
        cnt += 1
        ## 조건식을 생성 i와 j를 나눈 나머지 값이 0인가?
        if i % j == 0:
            # 조건식이 참이라면 check를 False 변경
            check = False
            # break
    # 조건문을 생성 check가 참인가?
    if check:
        # 참인 경우 i를 출력
        print(i, end='  ')
print("")
print(cnt)
```
